## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?
R- O canal que gera mais receita é aquele com maior volume de vendas, indicando maior participação no faturamento total da operação.

2. Qual canal gera maior eficiência?
R- O canal mais eficiente é o que apresenta maior margem de lucro, pois consegue gerar mais retorno proporcional sobre a receita, mesmo que o volume seja menor.

3. Se você fosse gestor, priorizaria volume ou margem?
R- Como gestor, eu não priorizaria apenas volume ou margem isoladamente. O ideal seria buscar equilíbrio: manter canais de alto volume para escala, mas focar na expansão dos canais mais eficientes, que geram maior rentabilidade.

```python
canal = df.groupby('canal_venda').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'count'
})

canal['margem'] = canal['lucro'] / canal['receita']
canal['ticket_medio'] = canal['receita'] / canal['pedido_id']

canal.sort_values('receita', ascending=False)
```

### Insight obrigatório
Escreva 1 ou 2 frases interpretando sua tabela, no estilo executivo:
- não apenas “o Online foi maior”
- mas o que isso **significa para a decisão**

R- A análise indica que o canal líder em receita não é necessariamente o mais eficiente em geração de lucro. Dessa forma, a estratégia deve priorizar não apenas volume de vendas, mas também a expansão dos canais com maior margem, visando aumento da rentabilidade global.

## 8. Análise por categoria

### Pergunta de negócio
**Qual categoria é mais lucrativa e qual parece mais eficiente?**

### Tarefa
Crie uma tabela agregada por `categoria` com:
- Receita Total
- Lucro Total
- Margem
- Ticket Médio

### Depois responda
1. A categoria de maior receita também é a de maior margem?
R- A categoria líder em receita não necessariamente apresenta a maior margem, evidenciando que volume de vendas não é sinônimo de eficiência.

2. Existe alguma categoria que parece vender bem, mas ser pouco eficiente?
R- Algumas categorias demonstram forte desempenho em faturamento, porém com margens reduzidas, indicando baixa eficiência operacional. Nesses casos, recomenda-se revisar custos, precificação ou até reposicionamento estratégico.

```python
categoria = df.groupby('categoria').agg({
    'receita': 'sum',
    'lucro': 'sum'
})

categoria['margem'] = categoria['lucro'] / categoria['receita']
categoria['ticket_medio'] = df.groupby('categoria')['receita'].mean()

categoria.sort_values('receita', ascending=False)
```

### Tarefa

- Receita Total

### Depois responda
1. Quais UFs concentram maior receita?

2. Existe concentração regional?
R- Observa-se uma concentração regional relevante, o que pode representar tanto uma vantagem competitiva quanto um risco de dependência geográfica.
3. Se fosse necessário expandir ou reforçar logística, por onde você começaria?
R- Estratégicamente, seria recomendável reforçar a logística nas regiões com maior demanda para ganho de eficiência, enquanto se explora expansão em UFs com menor participação, buscando diversificação e crescimento sustentável.


```python
uf = df.groupby('uf').agg({
    'receita': 'sum',
    'lucro': 'sum'
})

total_receita = df['receita'].sum()
uf['participacao'] = uf['receita'] / total_receita
uf['ticket_medio'] = df.groupby('uf')['receita'].mean()

uf.sort_values('receita', ascending=False)
```
