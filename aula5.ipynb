## 1. Preparação do ambiente

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Leitura da base

```python
df = pd.read_csv('vendas_brasil_clean_aula5.csv')
df.head()
df.shape
df.info()
```

## 3. Traduzindo a pergunta em gráfico
R- Pergunta do gestor: Qual canal gera mais receita?
R- Variáveis: canal_venda, receita
R- Tipo: barras
R- Biblioteca: matplotlib

R- Pergunta do gestor: Como a receita evolui?
R- Variáveis: data_venda, receita
R- Tipo: linha
R- Biblioteca: matplotlib

R- Pergunta: Margem varia por segmento?
R- Variáveis: segmento, margem_lucro
R- Tipo: boxplot
R- Biblioteca: seaborn

R- Pergunta: Receita x lucro?
R- Variáveis: receita, lucro
R- Tipo: scatter
R- Biblioteca: seaborn

## 4. Gráfico de barras

```python
canal = df.groupby('canal_venda')['receita'].sum().sort_values()
canal.plot(kind='barh')
plt.title('Canal concentra receita')
plt.show()
```

R- Canal líder concentra receita.
R- Estratégia: otimizar outros canais.

## 5. Linha

```python
df['data_venda']=pd.to_datetime(df['data_venda'])
df['mes']=df['data_venda'].dt.to_period('M')
tempo=df.groupby('mes')['receita'].sum()
tempo.plot()
plt.title('Receita ao longo do tempo')
plt.show()
```

R- Existe sazonalidade.
R- Planejar picos de venda.

## 6. Boxplot

```python
sns.boxplot(x='segmento', y='margem_lucro', data=df)
plt.show()
```

R- Segmentos variam em estabilidade.
R- Monitorar os mais instáveis.

## 7. Scatter

```python
sns.scatterplot(x='receita', y='lucro', data=df, alpha=0.5)
plt.show()
```

R- Relação positiva com exceções.
R- Investigar anomalias.
