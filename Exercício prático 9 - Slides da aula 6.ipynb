# Atividade Prática — Aula 6: Visualização Interativa com Plotly


## 1. Preparação do ambiente

```python
import pandas as pd
import plotly.express as px
```

## 2. Leitura da base

```python
df = pd.read_csv('vendas_brasil_clean_aula5.csv')
df.head()
df.info()
```

## 3. Pergunta → gráfico
R- Pergunta: Qual canal gera mais receita?
R- Variáveis: canal_venda, receita
R- Tipo: barras interativas

R- Pergunta: Como a receita evolui?
R- Variáveis: data_venda, receita
R- Tipo: linha


## 4. Gráfico de barras interativo

```python
canal = df.groupby('canal_venda')['receita'].sum().reset_index()
fig = px.bar(canal, x='receita', y='canal_venda', orientation='h', title='Receita por canal')
fig.show()
```

R- Canal líder concentra receita.
R- Ação: melhorar canais menores.

## 5. Linha interativa

```python
df['data_venda']=pd.to_datetime(df['data_venda'])
df['mes']=df['data_venda'].dt.to_period('M').astype(str)
tempo=df.groupby('mes')['receita'].sum().reset_index()
fig = px.line(tempo, x='mes', y='receita', title='Receita ao longo do tempo')
fig.show()
```

R- Existe sazonalidade.
R- Planejar picos.

## 6. Scatter interativo

```python
fig = px.scatter(df, x='receita', y='lucro', title='Receita vs Lucro')
fig.show()
```

R- Relação positiva.
R- Investigar outliers.

R- Gráficos interativos facilitam análise.
R- Insights mais rápidos.
